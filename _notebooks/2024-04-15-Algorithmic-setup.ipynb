## Bubble Sort
Checking one element with each element in the list

```java
import java.util.List;
import java.util.LinkedList;
import java.util.function.Consumer;

public class Flower implements Comparable<Flower> {
    private String name;
    private double height; // height in centimeters

    public Flower(String name, double height) {
        this.name = name;
        this.height = height;
    }

    public String getName() {
        return name;
    }

    public double getHeight() {
        return height;
    }

    public void setName(String name) {
        this.name = name;
    }

    public void setHeight(double height) {
        this.height = height;
    }

    @Override
    public String toString() {
        return "Flower{" +
               "name='" + name + '\'' +
               ", height=" + height + " cm" +
               '}';
    }

    public String toJSON() {
        return "{\"name\": \"" + name + "\", \"height\": " + height + "}";
    }

    @Override
    public int compareTo(Flower other) {
        return Double.compare(this.height, other.height);
    }

    // Sorting methods
    public static void bubbleSort(List<Flower> data) {
        boolean swapped;
        for (int i = 0; i < data.size() - 1; i++) {
            swapped = false;
            for (int j = 0; j < data.size() - 1 - i; j++) {
                if (data.get(j).compareTo(data.get(j + 1)) > 0) {
                    Flower temp = data.get(j);
                    data.set(j, data.get(j + 1));
                    data.set(j + 1, temp);
                    swapped = true;
                }
            }
            if (!swapped) break;
        }
    }

    public static void insertionSort(List<Flower> data) {
        for (int i = 1; i < data.size(); i++) {
            Flower key = data.get(i);
            int j = i - 1;
            while (j >= 0 && data.get(j).compareTo(key) > 0) {
                data.set(j + 1, data.get(j));
                j--;
            }
            data.set(j + 1, key);
        }
    }

    public static void mergeSort(List<Flower> data) {
        List<Flower> sortedList = mergeSortRec(new LinkedList<>(data));
        for (int i = 0; i < data.size(); i++) {
            data.set(i, sortedList.get(i));
        }
    }

    private static List<Flower> mergeSortRec(List<Flower> list) {
        if (list.size() <= 1) {
            return list;
        }
        int mid = list.size() / 2;
        List<Flower> left = mergeSortRec(new LinkedList<>(list.subList(0, mid)));
        List<Flower> right = mergeSortRec(new LinkedList<>(list.subList(mid, list.size())));
        return merge(left, right);
    }

    private static List<Flower> merge(List<Flower> left, List<Flower> right) {
        List<Flower> merged = new LinkedList<>();
        int leftIndex = 0, rightIndex = 0;
        while (leftIndex < left.size() && rightIndex < right.size()) {
            if (left.get(leftIndex).compareTo(right.get(rightIndex)) <= 0) {
                merged.add(left.get(leftIndex++));
            } else {
                merged.add(right.get(rightIndex++));
            }
        }
        merged.addAll(left.subList(leftIndex, left.size()));
        merged.addAll(right.subList(rightIndex, right.size()));
        return merged;
    }

    public static void selectionSort(List<Flower> data) {
        for (int i = 0; i < data.size() - 1; i++) {
            int minIndex = i;
            for (int j = i + 1; j < data.size(); j++) {
                if (data.get(j).compareTo(data.get(minIndex)) < 0) {
                    minIndex = j;
                }
            }
            if (minIndex != i) {
                Flower temp = data.get(i);
                data.set(i, data.get(minIndex));
                data.set(minIndex, temp);
            }
        }
    }

    public static void quickSort(List<Flower> data) {
        quickSortRec(data, 0, data.size() - 1);
    }

    private static void quickSortRec(List<Flower> data, int low, int high) {
        if (low < high) {
            int pi = partition(data, low, high);
            quickSortRec(data, low, pi - 1);
            quickSortRec(data, pi + 1, high);
        }
    }

    private static int partition(List<Flower> data, int low, int high) {
        Flower pivot = data.get(high);
        int i = (low - 1);
        for (int j = low; j < high; j++) {
            if (data.get(j).compareTo(pivot) <= 0) {
                i++;
                Flower temp = data.get(i);
                data.set(i, data.get(j));
                data.set(j, temp);
            }
        }
        Flower temp = data.get(i + 1);
        data.set(i + 1, data.get(high));
        data.set(high, temp);
        return i + 1;
    }

    public static void main(String[] args) {
        List<Flower> flowersDataset = new LinkedList<>();
        Random random = new Random();
        for (int i = 0; i < 1000; i++) {
            flowersDataset.add(new Flower("Flower" + i, random.nextDouble() * 100));
        }

        List<Flower> flowers = new LinkedList<>(flowersDataset);

        Consumer<List<Flower>> displayFlowers = (list) -> {
            list.stream().limit(5).forEach(flower -> System.out.println(flower.toJSON()));
        }; // The .stream method creats a sequential stream of the objects in the List folowers. It allows other functions to take place on these "stream" of elements like limit (limiting the stream to the first 5 elements)

        long startTime, endTime;

        // Insertion Sort
        startTime = System.currentTimeMillis();
        Flower.insertionSort(flowers);
        endTime = System.currentTimeMillis();
        System.out.println("Elapsed time for Insertion Sort: " + (endTime - startTime) + " ms");
        displayFlowers.accept(flowers);


        // Reset and sort again for each type
        flowers = new LinkedList<>(flowersDataset);
        startTime = System.currentTimeMillis();
        Flower.bubbleSort(flowers);
        endTime = System.currentTimeMillis();
        System.out.println("Elapsed time for Bubble Sort: " + (endTime - startTime) + " ms");
        displayFlowers.accept(flowers);


        flowers = new LinkedList<>(flowersDataset);
        startTime = System.currentTimeMillis();
        Flower.mergeSort(flowers);
        endTime = System.currentTimeMillis();
        System.out.println("Elapsed time for Merge Sort: " + (endTime - startTime) + " ms");
        displayFlowers.accept(flowers);


        flowers = new LinkedList<>(flowersDataset);
        startTime = System.currentTimeMillis();
        Flower.quickSort(flowers);
        endTime = System.currentTimeMillis();
        System.out.println("Elapsed time for Quick Sort: " + (endTime - startTime) + " ms");
        displayFlowers.accept(flowers);


        flowers = new LinkedList<>(flowersDataset);
        startTime = System.currentTimeMillis();
        Flower.selectionSort(flowers);
        endTime = System.currentTimeMillis();
        System.out.println("Elapsed time for Selection Sort: " + (endTime - startTime) + " ms");
        displayFlowers.accept(flowers);

    }
}

Flower.main(null)
```

Code explain
Java class named AllSorts that implements several classic sorting algorithms. Each algorithm is applied to sort a list of data that the class maintains. The class works with generic data types (T), constrained to types that implement the Comparable interface, meaning the data elements can be compared to one another.

Here’s a breakdown of each component of the class:

Imports:

java.util.List: Interface representing ordered collections (lists) of elements.
java.util.ArrayList: Implementation of List backed by an array.
java.util.LinkedList: Implementation of List using a doubly-linked list.
java.util.Random: Utility class for generating random values.
Class Definition:

AllSorts<T extends Comparable<T>>: The class is generic, accepting only types T that can be compared (i.e., they implement the Comparable interface). This allows elements to be ordered.
Constructor:

AllSorts(List<T> data): Constructs an instance of AllSorts with an initial list of data.
Sorting Methods: Each method sorts the internal list data using different algorithms.

Bubble Sort (bubbleSort): A simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The algorithm stops early if no swaps were needed on a new pass.
Insertion Sort (insertionSort): Builds the final sorted array one item at a time, placing each new item into its correct position relative to already sorted items.
Merge Sort (mergeSort): A divide and conquer algorithm that splits the list into two halves, recursively sorts both halves, and then merges the two sorted halves into a single sorted list.
Selection Sort (selectionSort): Selects the minimum element from the unsorted portion of the list, and then swaps it with the element at the beginning of the unsorted portion, progressively moving the boundary of the unsorted portion.
Quick Sort (quickSort): Another divide and conquer algorithm that partitions the list around a pivot element, sorting elements around the pivot, and then recursively applying the process to the sub-lists.
Private Helper Methods for mergeSort and quickSort:

mergeSortRec and merge: Recursively split and merge for the merge sort.
quickSortRec and partition: Recursive quick sort implementation and its partitioning logic.
Each sorting method alters the data list in-place (i.e., it modifies the original list instead of creating a sorted copy). These implementations are typical examples of their respective algorithms and can be used to sort any list of objects that can be compared to one another, such as integers, strings, or custom objects implementing the Comparable interface.

```java
package com.nighthawk.spring_portfolio.mvc.stocks.sorting;

import java.util.List;

import org.springframework.util.StopWatch;

public class Descending {

  StopWatch sw = new StopWatch();
  int comparisons;

  public SortedData selectionSort(List<MonthlyStocks> monthlyStocksList) { // here monthlystocks is a database class we have that holds the data on the stocks (monthly)
    int n = monthlyStocksList.size();

    sw.start();
    for (int i = 0; i < n - 1; i++) {
      int maxIndex = i;

      for (int j = i + 1; j < n; j++) { 
        if (compareStrings(monthlyStocksList.get(j).getHigh(), monthlyStocksList.get(maxIndex).getHigh()) > 0) {
          maxIndex = j;
        }
        comparisons++;
      }

      // Swap elements
      MonthlyStocks temp = monthlyStocksList.get(maxIndex);
      monthlyStocksList.set(maxIndex, monthlyStocksList.get(i));
      monthlyStocksList.set(i, temp);
    }
    sw.stop();

    return new SortedData(comparisons, sw.getTotalTimeMillis(), monthlyStocksList);
  }

}
```
