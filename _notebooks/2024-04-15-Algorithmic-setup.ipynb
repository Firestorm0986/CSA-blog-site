```java
import java.util.ArrayList;

public class Flower implements Comparable<Flower> {
    private String name;
    private double height; // height in centimeters

    public Flower(String name, double height) {
        this.name = name;
        this.height = height;
    }

    public String getName() {
        return name;
    }

    public double getHeight() {
        return height;
    }

    public void setName(String name) {
        this.name = name;
    }

    public void setHeight(double height) {
        this.height = height;
    }

    @Override
    public String toString() {
        return "Flower{" +
               "name='" + name + '\'' +
               ", height=" + height + " cm" +
               '}';
    }

    @Override
    public int compareTo(Flower other) {
        return Double.compare(this.height, other.height);
    }
}
```

## Bubble Sort
Checking one element with each element in the list

```java
import java.util.List;
import java.util.ArrayList;
import java.util.LinkedList;
import java.util.Random;

public class AllSorts<T extends Comparable<T>> {
    private List<T> data;

    public AllSorts(List<T> data) {
        this.data = data;
    }

    public void bubbleSort() {
        boolean swapped;
        for (int i = 0; i < data.size() - 1; i++) {
            swapped = false;
            for (int j = 0; j < data.size() - 1 - i; j++) {
                if (data.get(j).compareTo(data.get(j + 1)) > 0) {
                    T temp = data.get(j);
                    data.set(j, data.get(j + 1));
                    data.set(j + 1, temp);
                    swapped = true;
                }
            }
            if (!swapped) break;
        }
    }

    public void insertionSort() {
        for (int i = 1; i < data.size(); i++) {
            T key = data.get(i);
            int j = i - 1;

            while (j >= 0 && data.get(j).compareTo(key) > 0) {
                data.set(j + 1, data.get(j));
                j--;
            }
            data.set(j + 1, key);
        }
    }

    public void mergeSort() {
        List<T> sortedList = mergeSortRec(new LinkedList<>(data));
        for (int i = 0; i < data.size(); i++) {
            data.set(i, sortedList.get(i));
        }
    }

    private List<T> mergeSortRec(List<T> list) {
        if (list.size() <= 1) {
            return list;
        }
        int mid = list.size() / 2;
        List<T> left = mergeSortRec(new LinkedList<>(list.subList(0, mid)));
        List<T> right = mergeSortRec(new LinkedList<>(list.subList(mid, list.size())));
        return merge(left, right);
    }

    private List<T> merge(List<T> left, List<T> right) {
        List<T> merged = new LinkedList<>();
        int leftIndex = 0, rightIndex = 0;
        while (leftIndex < left.size() && rightIndex < right.size()) {
            if (left.get(leftIndex).compareTo(right.get(rightIndex)) <= 0) {
                merged.add(left.get(leftIndex++));
            } else {
                merged.add(right.get(rightIndex++));
            }
        }
        merged.addAll(left.subList(leftIndex, left.size()));
        merged.addAll(right.subList(rightIndex, right.size()));
        return merged;
    }

    public void selectionSort() {
        for (int i = 0; i < data.size() - 1; i++) {
            int minIndex = i;
            for (int j = i + 1; j < data.size(); j++) {
                if (data.get(j).compareTo(data.get(minIndex)) < 0) {
                    minIndex = j;
                }
            }
            if (minIndex != i) {
                T temp = data.get(i);
                data.set(i, data.get(minIndex));
                data.set(minIndex, temp);
            }
        }
    }

    public void quickSort() {
        quickSortRec(0, data.size() - 1);
    }

    private void quickSortRec(int low, int high) {
        if (low < high) {
            int pi = partition(low, high);
            quickSortRec(low, pi - 1);
            quickSortRec(pi + 1, high);
        }
    }

    private int partition(int low, int high) {
        T pivot = data.get(high);
        int i = (low - 1);
        for (int j = low; j < high; j++) {
            if (data.get(j).compareTo(pivot) <= 0) {
                i++;
                T temp = data.get(i);
                data.set(i, data.get(j));
                data.set(j, temp);
            }
        }
        T temp = data.get(i + 1);
        data.set(i + 1, data.get(high));
        data.set(high, temp);
        return i + 1;
    }
}
```

```java
import java.util.List;
import java.util.LinkedList;
import java.util.Random;

public class FlowerDatabase {
    public static void main(String[] args) {
        List<Flower> flowersDataset = new LinkedList<>();
        Random random = new Random();
        for (int i = 0; i < 1000; i++) {
            flowersDataset.add(new Flower("Flower" + i, random.nextDouble() * 100));
        }

        // Sort using Insertion Sort
        List<Flower> flowersI = new LinkedList<>(flowersDataset);
        AllSorts<Flower> sorterI = new AllSorts<>(flowersI);
        long startTimeI = System.currentTimeMillis();
        sorterI.insertionSort();
        long endTimeI = System.currentTimeMillis();
        System.out.println("Elapsed time for Insertion Sort: " + (endTimeI - startTimeI) + " ms");

        // Sort using Bubble Sort
        List<Flower> flowersB = new LinkedList<>(flowersDataset);
        AllSorts<Flower> sorterB = new AllSorts<>(flowersB);
        long startTimeB = System.currentTimeMillis();
        sorterB.bubbleSort();
        long endTimeB = System.currentTimeMillis();
        System.out.println("Elapsed time for Bubble Sort: " + (endTimeB - startTimeB) + " ms");

        // Sort using Merge Sort
        List<Flower> flowersM = new LinkedList<>(flowersDataset);
        AllSorts<Flower> sorterM = new AllSorts<>(flowersM);
        long startTimeM = System.currentTimeMillis();
        sorterM.mergeSort();
        long endTimeM = System.currentTimeMillis();
        System.out.println("Elapsed time for Merge Sort: " + (endTimeM - startTimeM) + " ms");

        // Sort using Quick Sort
        List<Flower> flowersQ = new LinkedList<>(flowersDataset);
        AllSorts<Flower> sorterQ = new AllSorts<>(flowersQ);
        long startTimeQ = System.currentTimeMillis();
        sorterQ.quickSort();
        long endTimeQ = System.currentTimeMillis();
        System.out.println("Elapsed time for Quick Sort: " + (endTimeQ - startTimeQ) + " ms");

        // Sort using Selection Sort
        List<Flower> flowersS = new LinkedList<>(flowersDataset);
        AllSorts<Flower> sorterS = new AllSorts<>(flowersS);
        long startTimeS = System.currentTimeMillis();
        sorterS.selectionSort();
        long endTimeS = System.currentTimeMillis();
        System.out.println("Elapsed time for Selection Sort: " + (endTimeS - startTimeS) + " ms");

        // Verification (Optional)
        // Check if all lists are sorted the same way (you might need to compare all the lists)
        boolean correctSort = true;
        for (int j = 0; j < flowersDataset.size(); j++) {
            if (!flowersI.get(j).equals(flowersB.get(j)) ||
                !flowersB.get(j).equals(flowersM.get(j)) ||
                !flowersM.get(j).equals(flowersS.get(j)) ||
                !flowersS.get(j).equals(flowersQ.get(j))) {
                correctSort = false;
                break;
            }
        }

        if (correctSort) {
            System.out.println("All sorted lists are the same.");
        } else {
            System.out.println("There is a discrepancy between the sorted lists.");
        }
    }
}

FlowerDatabase.main(null)
```

Seeing as there are so many possibilities we need a common sort class that sorts (as most are about integers and money made we can keep the datatype to Double)

Example: The code below is for sorting based on descending order (big to small). As this type of sort will be mostly used I have this set up. This is also set up to check the time taken. 

```java
package com.nighthawk.spring_portfolio.mvc.stocks.sorting;

import java.util.List;

import org.springframework.util.StopWatch;

public class Descending {

  StopWatch sw = new StopWatch();
  int comparisons;

  public SortedData selectionSort(List<MonthlyStocks> monthlyStocksList) { // here monthlystocks is a database class we have that holds the data on the stocks (monthly)
    int n = monthlyStocksList.size();

    sw.start();
    for (int i = 0; i < n - 1; i++) {
      int maxIndex = i;

      for (int j = i + 1; j < n; j++) { 
        if (compareStrings(monthlyStocksList.get(j).getHigh(), monthlyStocksList.get(maxIndex).getHigh()) > 0) {
          maxIndex = j;
        }
        comparisons++;
      }

      // Swap elements
      MonthlyStocks temp = monthlyStocksList.get(maxIndex);
      monthlyStocksList.set(maxIndex, monthlyStocksList.get(i));
      monthlyStocksList.set(i, temp);
    }
    sw.stop();

    return new SortedData(comparisons, sw.getTotalTimeMillis(), monthlyStocksList);
  }

}
```
